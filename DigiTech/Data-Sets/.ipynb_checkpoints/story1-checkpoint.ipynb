```python
#importing necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
#Loading the dataset
data = pd.read_csv('googlePlayStore.csv')
```

```python
# Handle missing values for Rating
data['Rating'] = data['Rating'].fillna(data['Rating'].mean())
```

```python
# Handle non-numeric and missing values for Reviews
data['Reviews'] = pd.to_numeric(data['Reviews'], errors='coerce').fillna(0).astype(int)
```

```python
# Handle non-numeric values for Installs
# Remove non-numeric entries first
data = data[data['Installs'].str.contains(r'^\d+[,+]?\d*$', na=False)]
```

```python
# Remove commas and plus signs, then convert to integers
data['Installs'] = data['Installs'].str.replace('[+,]', '', regex=True).astype(int)
```

```python
# Convert categorical variables to dummy/indicator variables
data = pd.get_dummies(data, columns=['Category', 'Type', 'Content Rating', 'Genres'])
```
